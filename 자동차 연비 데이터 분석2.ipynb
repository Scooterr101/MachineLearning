```python
# 오리진이 각각의 나라를 의미

# origin : 제조국 1 USA 2 EU 3 JPN
```

```python
df.groupby('origin').size()
```

```python
# 벨류카운트랑 똑같은 값이 나옴
```

```python
df_org = pd.DataFrame()
df_org['count'] = df.groupby('origin').size()
```

```python
df_org['count']
```

```python
df_org['count'].plot(kind='pie')
```

```python
df_org.index = ["USA", "EU", "JAPAN"]
df_org["count"].plot(kind='pie',
                     figsize=(14, 6),
                     startangle=10,
                     colors=["chocolate","bisque","cadetblue"],
                    autopct = '%1.1f%%')

plt.title("Model Origin", size = 20)
plt.axis('equal')
plt.legend(labels=df_org.index, loc = 'best') # 베스트는 제일 좋은 자리 찾아줌
plt.show()
```

```python
fig = plt.figure(figsize=(15,5)) # 도화지를 하나 만듬

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

ax1.boxplot(x=df[df['origin']==1]['mpg'])
ax2.boxplot(x=df[df['origin']==2]['mpg'])
ax3.boxplot(x=df[df['origin']==3]['mpg'])

plt.show()
```
